# Facebook V: Predicting Check Ins
## Identify the correct place for check ins
The goal of this competition is to predict which place a person would like to check in to. For the purposes of this competition, Facebook created an artificial world consisting of more than 100,000 places located in a 10 km by 10 km square. For a given set of coordinates, your task is to return a ranked list of the most likely places. Data was fabricated to resemble location signals coming from mobile devices, giving you a flavor of what it takes to work with real data complicated by inaccurate and noisy values. Inconsistent and erroneous location data can disrupt experience for services like Facebook Check In.

There is evidence that suggests that the timestamp is in minutes:
https://www.kaggle.com/andersonk/facebook-v-predicting-check-ins/timestamps-structure/comments.
Let's calculate the hour, day, week, month and yeark breaks-ups and add them as features.

```python
df_train['hour'] = df_train.time // 60 % 24 + 1
df_train['day'] = df_train.time // (60*24) % 7 + 1
df_train['week'] = df_train.time // (60*24*7) % 52 + 1
df_train['month'] = df_train.time // (60*24*30) % 12 + 1
df_train['year'] = df_train.time // (60*24*365) + 1

df_train['hours'] = df_train.time // 60 + 1
df_train['days'] = df_train.time // (60*24) + 1
df_train['weeks'] = df_train.time // (60*24*7) + 1
df_train['months'] = df_train.time // (60*24*30) + 1

df_train.head()
```

Let's add the rise, fall or peak as features to the dataset and re-plot this with color labels

```python
def apply_accu_class(x):
    rise_ranges = ((50, 90), (130, 160), (220, 310), (380, 600))
    fall_ranges = ((0, 40), (110, 120), (190, 210), (320, 370), )
    peak_ranges = ((90, 110), (160, 190), (310, 320), )
    trough_ranges = ((40, 50), (120, 130), (210, 220), (370, 380))
    class_dict = {rise_ranges: 1, fall_ranges: -1, peak_ranges: 2, trough_ranges: -2}
    for rang in [rise_ranges, fall_ranges, peak_ranges, trough_ranges]:
        for a, b in rang:
            if a <= x < b:
                return class_dict[rang]

df_train['accu_class'] = df_train.days.apply(apply_accu_class)
```

```python
df_acc_time = df_train.groupby(['days']).mean().reset_index()
plt.scatter(df_acc_time.days, df_acc_time.accuracy, c=df_acc_time.accu_class)
plt.rcParams["figure.figsize"] = (10, 10)
plt.show()
```

It seems difficult to draw any concrete conclusions about the rise and fall in accuracy. Perhaps these checkins are from locations that are that are only open at certain times during the year? Maybe these seasonal locations are close to cell towers and therefore have a higher accuracy? There are no leading theories about accuracy. I am hoping that adding log10, log2 and the accuracy class parameter will help the ML algorithm discover latent meaning in the accuracy field.
