```python
# There generally seems to be a lot more variation in the X co-ordinates than in the Y coordinates.
# What we'd like to do is some sort of sliding window sampling of the training dataset based on the X,Y coordinates.
# There are a total of 29118021 data points and 108390 distinct places (N) to predict on
# We will then train 'K' separate models which would be at max a 'M' class classification model where M << N
```

```python
class PredictionModel():
    def __init__(self, df, xsize=1, ysize=0.5, xslide=0.5, yslide=0.25, xcol='x', ycol='y'):
        self.df = df
        self.xsize = xsize
        self.ysize = ysize
        self.xslide = xslide
        self.yslide = yslide
        self.xcol = xcol
        self.ycol = ycol
        
        self.windows = self.generate_windows()
        self.slices = self.slice_df()
    
    def frange(self, x, y, jump):
        while x < y:
            yield x
            x += jump
        yield y

    def generate_windows(self):  
        ranges = []
        result = []
        
        xmin, xmax = self.df.x.min(), self.df.x.max()
        ymin, ymax = self.df.y.min(), self.df.y.max()
        xranges = list(self.frange(xmin, xmax-self.xsize, self.xslide))
        yranges = list(self.frange(ymin, ymax-self.ysize, self.yslide))
        ylen = len(yranges)
        for x in xranges:
            subrange = [x] * ylen
            ranges.extend(zip(subrange, yranges))

        for x1, y1 in ranges:
            x2, y2 = x1 + self.xsize, y1 + self.ysize
            result.append(((x1, y1), (x1+self.xsize, y1+self.ysize)))
        
        return result[:2]

    def slice_df(self):
        slices = {}
        for window in self.windows:
            slices[window] = ModelStore(self.df, window, self.xcol, self.ycol)

        return slices
    
    def find_best_window(self, x, y):
        xs = max(0, x - (self.xsize/2.0))
        ys = max(0, y - (self.ysize/2.0))
        x0, y0 = 0, 0

        while x0 < xs:
            x0 += self.xslide
        while y0 < ys:
            y0 += self.yslide
            
        x1, x2 = x0, x0 + self.xsize
        y1, y2 = y0, y0 + self.ysize
    
        try:
            assert x1 <= x <= x2
            assert y1 <= y <= y2
        except:
            import pdb; pdb.set_trace()

        return ((x1, y1), (x2, y2))
        
    def predict(self, df):
        pass

    
class ModelStore():
    def __init__(self, df, window, xcol, ycol):
        self.window = window
        (self.x1, self.y1), (self.x2, self.y2) = self.window 
        self.df = df[(df[xcol] >= self.x1) & (df[xcol] <= self.x2) & (df[ycol] >= self.y1) & (df[ycol] <= self.y2)]
        self.unique_place_count = len(self.df.place_id.unique())
        self.model = None
        self.df['hours'] = self.df.time / 60.0
        self.df['days'] = self.df.time / (60*24)
        self.df['hours_cycle'] = self.df.hours % 24
        self.df['days_cycle'] = self.df.days % 7
        
    def train(self):
        from sklearn.ensemble import RandomForestClassifier
        self.model = RandomForestClassifier(n_estimators=5)  # x, y, accuracy, hours_cycle, days_cycle
        self.train_df = self.df.sort_values('row_id')[['x', 'y', 'accuracy', 'hours_cycle', 'days_cycle']]
        self.values = self.df.sort_values('row_id')['place_id']
        self.model.fit(self.train_df, self.values)
```

```python
df_under_2 = df_train[(df_train.x <= 2.0) & (df_train.y <= 2.0)]
train, test = train_test_split(df_under_2, test_size = 0.2)

model = PredictionModel(df=train)
```
